```python
import math
import scipy.stats as ss
from scipy.stats.stats import spearmanr
from scipy.stats.stats import pearsonr
```

## Exercise 1

```python
#provided:
sumx=1190.0
sumy=248.6
sumx2=95098
sumy2=4161.1
sumxy=19862.6
```

```python
#Chirps/Second 
y=[14.4,14.7,15.0,15.4,15.5,15.7,16.0,16.1,16.3,17.0,17.1,17.2,18.4,19.8,20.0]
```

```python
#Temperature
x = [76.3, 69.7, 79.6, 69.4, 75.2, 71.5, 71.6, 80.5, 83.3, 83.5, 80.6, 82.6, 84.3, 93.3, 88.6]
```

#### a) Formulate the hypothesis testing that the two variables are uncorrelated against they are positively correlated at 1% significant level

```python
# Using stats libraries
print("The Pearson coeff. is: ",pearsonr(x,y)[0])

#####################################
# By calculation
Sxx = sumx2 - (sumx)**2/len(x)
Syy = sumy2 - (sumy)**2/len(x)
Sxy = sumxy - (sumx*sumy)/len(x)

r = Sxy/math.sqrt(Sxx*Syy)
print("By calculation, the Pearson coeff. is: ",r) 
```

From the Table of Critical Values for Pearson’s r  ( 1% , 13 DoF) point is 0.592.
Therefore at 1% significance level there is strong evidence against null hypothesis, which shows that there is a strong positive correlation between number of chirps and temperature.

#### b) Calculate the Spearman's rank correlation coefficient and test for 1% significant level for the hypothesis testing

```python
# By calculation
rankx = list(ss.rankdata(x))                # rank the values
ranky = list(ss.rankdata(y))

n=len(rankx)
d=sum([((a-b)**2) for a,b in zip(rankx,ranky)])
rs = 1- (6*d)/(n*(n**2-1))
print ("By calculation, the Spearman's rank correlation coeff. is: ",rs)

##############################################
# Using stats libraries
print (spearmanr(x,y))
```

1% point 0.604 ⇒ Strong evidence against H0, which shows that there is a strong association between the number of chirps and temperature

## Exercise 2

```python
# Silver 
x = [29, 27, 17, 25, 8, 19, 11, 9, 4]
# Bronze 
y = [29, 22, 19, 33, 7, 14, 12, 11, 5]
```

```python
sum(y)
```

#### a) Formulate the hypothesis testing that the two variables are uncorrelated against they arepositively correlated at 2% significant level

```python
# Using stats libraries
print("Pearson r value is: ",pearsonr(x,y)[0])

#####################################
# By calculation
sumx=sum(x)
sumy=sum(y)
sumx2=sum([i**2 for i in x])
sumy2=sum([j**2 for j in y])
sumxy=sum([(x[k]*y[k]) for k in range(len(x))])

Sxx = sumx2 - (sumx)**2/len(x)
Syy = sumy2 - (sumy)**2/len(x)
Sxy = sumxy - (sumx*sumy)/len(x)

r = Sxy/math.sqrt(Sxx*Syy)
print("By calculation, the Pearson coeff. is: ",r) 
```

```python
# critical value for 2% significant level, 13 DoF is:
(0.750-0.666)/3*2+0.666
```

The critical value for (2% significant, 13 DoF) is 0.722. 
Since 0.913 > 0.722, we reject the null hypothesis in favor of alternative hypothesis

#### b) Calculate the Spearman's rank correlation coefficient and test for 2% significant level for thehypothesis testing

```python
# By calculation
rankx = list(ss.rankdata(x))
ranky = list(ss.rankdata(y))

n=len(x)
d=sum([((a-b)**2) for a,b in zip(rankx,ranky)])
rs = 1- (6*d)/(n*(n**2-1))
print ("By calculation, the Spearman's rank correlation coeff. is: ",rs)

##############################################
# Using stats libraries
spearmanr(x,y)
```
